```python
# Notebook 1: Ingesta del CSV crudo
# Lee el archivo desde el Volume y lo guarda como tabla "retail_raw"

# Leer la tabla existente
df = spark.table("workspace.default.retail_store_sales")

# Mostrar cuántas filas y columnas tiene
print(f"Filas leídas: {df.count()}")
print(f"Columnas: {len(df.columns)}")

# Vista previa de los primeros registros
df.show(5)
```

```python
# Guardar como tabla Bronze (capa raw en el catálogo)
df.write.mode("overwrite") \
  .option("delta.columnMapping.mode", "name") \
  .saveAsTable("workspace.default.retail_raw")

print("Tabla 'retail_raw' creada exitosamente.")
```
